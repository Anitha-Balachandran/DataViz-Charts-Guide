```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("train.csv")
df_class = df.groupby(['Survived', 'Pclass']).size().unstack()
fig, ax = plt.subplots(figsize=(10,6))
df_class.plot(kind='bar', stacked=True, ax=ax, color=["#f29a10", "#c43232","#3d85c6"])
ax.set_xlabel('Survived')
ax.set_ylabel('Number of Passengers')
ax.set_title('Stacked Bar Chart: Passenger Survival by Class')
plt.show()
print('Understanding Amounts:\n \n The data used in the above plot has two categories,(survived or not survived, and passenger class sub-categories).\n The number of passengers in each category is plotted on the y-axis.\n The above stacked bar chart allows us to compare the number of passengers who survived or did not survive within each passenger class.\n \n Observations:\n i) From the above chart, The number of third-class passengers was significantly higher than the number of first or second-class passengers.\n ii) First-class passengers had a higher survival rate than second or third-class passengers, and the majority of survivors were from first class.\n iii) Third-class passengers had a higher number of fatalities than first or second-class passengers, and the majority of fatalities were from third class.')
```

```python
import matplotlib.pyplot as plt
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temp = [20, 18, 22, 19, 25, 28, 23]
fig, ax = plt.subplots()
ax.plot(days, temp)
ax.set_xlabel('Day')
ax.set_ylabel('Temperature (C)')
ax.set_title('Line Graph: Daily Temperature')
plt.show()
print('Understanding x-y relationships: \n\n In the above line graph, the x-axis represents the days of the week and the y-axis represents the temperature. \n The line connecting the points shows the trend and relationship between the temperature over the course of the week. \n \n Observations: \n i) the temperature in the chart fluctuates throughout the week, with a peak on Saturday and a slight dip on Tuesday. \n ii) By using a line graph to represent this data, we can easily see the general trend in temperature over time.')
```

```python
import numpy as np
import matplotlib.pyplot as plt
temp = [21.2, 21.5, 21.1, 20.8, 20.9, 21.3, 22.1, 23.2, 23.5, 23.7, 23.9, 24.1, 23.8, 23.6]
mean = np.mean(temp)
std = np.std(temp)
sem = std / np.sqrt(len(temp))
conf_level = 0.95
z_score = 1.96
upper = mean + z_score * sem
lower = mean - z_score * sem
x = np.arange(len(temp))
plt.plot(x, temp, 'o-', label='Temperature')
plt.fill_between(x, upper, lower, alpha=0.2, label='95% Confidence Band')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Temperature (C)')
plt.title('Confidence Band: Temperature Variation over Time')
plt.show()
print('Understanding of Uncertainty: \n \n The above plot gives us a visual representation of the variation and uncertainty in temperature data over time. \n Observation: \n i) we can observe that the temperature varies from day to day, and the graded confidence band gives us an idea of the uncertainty associated with the temperature.  \n  ii) the average temperature is highest in the middle of the week and lower on the weekends and the band widens throughout.')
```
